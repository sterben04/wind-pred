```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv('D:\ML\data\T1.csv')

dataset.drop('Theoretical_Power_Curve (KWh)',axis=1,inplace=True)
dataset.dtypes

```

```python
dataset['Date/Time']=pd.to_datetime(dataset['Date/Time'])
```

```python
dataset.dtypes
```

```python
data=dataset.copy()
data.info()
```

```python
#feature scaling

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(data[['Wind Speed (m/s)','Wind Direction (°)']])
```

```python

data.head()
```

```python
scaler.transform(data[['Wind Speed (m/s)',
                       'Wind Direction (°)']])
```

```python
data=pd.concat([data[['Date/Time','LV ActivePower (kW)']],pd.DataFrame(
    scaler.transform(data[['Wind Speed (m/s)','Wind Direction (°)']]),columns=['Wind Speed (m/s)','Wind Direction (°)'])],axis=1)
```

```python
data.head()
```

```python
y=data['LV ActivePower (kW)'].values
y



# print(np.shape(X))
# print(np.shape(y))

# X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
data.drop('LV ActivePower (kW)',axis=1,inplace=True)
data
```

```python
print(data.shape)
print(y.shape)
```

```python
data['Date/Time']=pd.to_numeric(data['Date/Time'])
```

```python
data.head()
```

```python
data.dtypes
```

```python
X_train,X_test,y_train,y_test=train_test_split(data,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=10,random_state=0)
from sklearn.model_selection import cross_val_score
regressor.fit(X_train,y_train)
train_score = cross_val_score(regressor,X_train,y_train,cv=5)
test_score = cross_val_score(regressor,X_test,y_test,cv=5)

print("train score :",np.mean(train_score))
print("test score :",np.mean(test_score))
```
